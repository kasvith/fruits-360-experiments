```python
# Import the `pyplot` module
import matplotlib.pyplot as plt 

# Make a histogram with 90 bins of the `labels` data
fig = plt.hist(labels, 90)
names = list(set(labels)).sort()
plt.xticks(names, rotation='vertical')
plt.rcParams["figure.figsize"] = (39, 9)

# Show the plot
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12, 12)

# random fruits from set
fruits = [300, 2250, 3650, 4000]

# Fill out the subplots with the random images and add shape, min and max values
for i in range(len(fruits)):
    plt.subplot(1, 4, i+1)
    plt.axis('off')
    plt.imshow(images[fruits[i]])
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    print("shape: {0}, min: {1}, max: {2}".format(images[fruits[i]].shape, 
                                                  images[fruits[i]].min(), 
                                                  images[fruits[i]].max()))
```
